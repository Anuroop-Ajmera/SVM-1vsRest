```python
import pandas as pd
import numpy as np
#import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
import matplotlib.pyplot as plt
```

### Step-1

```python
dataset = pd.read_csv('./letter-recognition.csv')
dataset.head(5)
```

```python
train_set = pd.DataFrame(data=None, columns = dataset.columns)
test_set = pd.DataFrame(data=None, columns = dataset.columns)
#train_set
```

### Step-2

```python
def train_test_splitting(full_dataframe:pd.DataFrame, tgt_class:str):
    global train_set
    global test_set
    filtered_dataframe = full_dataframe[full_dataframe.letter==tgt_class]
    train, test = train_test_split(filtered_dataframe, test_size = 0.3)
    train_set = pd.concat([train_set,train], axis=0)
    test_set = pd.concat([test_set,test], axis=0)
```

```python
dataset.letter.unique()
```

### Step-3

```python
for tgt_class in dataset.letter.unique():
    #print("target class is:",tgt_class)
    train_test_splitting(dataset, tgt_class)
```

### Step-4

```python
X_train = train_set.drop(['letter'], axis=1)
y_train = train_set['letter']
X_test = test_set.drop(['letter'], axis=1)
y_test = test_set['letter']
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import svm
from sklearn.svm import SVC
```

```python
print(dataset.min())
print(dataset.max())
```

### As we can see that the range of the all independent variables is between 0 and 15 so standardization is not required

```python
unique_letters = dataset['letter'].sort_values().unique()
#print(unique_letters)
accuracy_df = pd.DataFrame(unique_letters, columns = ['Class_label'])
accuracy_df['Linear_C1'] = np.NaN
accuracy_df['Linear_C5'] = np.NaN
accuracy_df['Linear_C10'] = np.NaN
accuracy_df['Polynomial_d2_C5'] = np.NaN
accuracy_df['Polynomial_d3_C5'] = np.NaN
accuracy_df['RBF_C5'] = np.NaN
accuracy_df
```

### 1-vs-rest SVM classifier for each class

```python
for i in range(0,len(accuracy_df['Class_label'])):
    class_label_iter = accuracy_df.iloc[i,0]

    svm_mod1 = svm.SVC(kernel='linear',C=1.0)
    svm_mod2 = svm.SVC(kernel='linear',C=5.0)
    svm_mod3 = svm.SVC(kernel='linear',C=10.0)
    svm_mod4 = svm.SVC(kernel='poly',C=5.0,degree=2)
    svm_mod5 = svm.SVC(kernel='poly',C=5.0,degree=3)
    svm_mod6 = svm.SVC(kernel='rbf',C=5.0)
    
    ####################################################
    svm_mod1.fit(X_train,y_train)
    y_pred1 = svm_mod1.predict(X_test)
    accuracy_mod1 = metrics.accuracy_score(y_test, y_pred1)
    print("Accuracy calculated for Class_label "+ class_label_iter + ": Model 1") 

    ####################################################
    svm_mod2.fit(X_train,y_train)
    y_pred2 = svm_mod2.predict(X_test)
    accuracy_mod2 = metrics.accuracy_score(y_test, y_pred2)
    print("Accuracy calculated for Class_label "+ class_label_iter + ": Model 2") 

    ####################################################
    svm_mod3.fit(X_train,y_train)
    y_pred3 = svm_mod3.predict(X_test)
    accuracy_mod3 = metrics.accuracy_score(y_test, y_pred3)
    print("Accuracy calculated for Class_label "+ class_label_iter + ": Model 3") 

    ####################################################
    svm_mod4.fit(X_train,y_train)
    y_pred4 = svm_mod4.predict(X_test)
    accuracy_mod4 = metrics.accuracy_score(y_test, y_pred4)
    print("Accuracy calculated for Class_label "+ class_label_iter + ": Model 4") 

    ####################################################
    svm_mod5.fit(X_train,y_train)
    y_pred5 = svm_mod5.predict(X_test)
    accuracy_mod5 = metrics.accuracy_score(y_test, y_pred5)
    print("Accuracy calculated for Class_label "+ class_label_iter + ": Model 5") 

    ####################################################
    svm_mod6.fit(X_train,y_train)
    y_pred6 = svm_mod6.predict(X_test)
    accuracy_mod6 = metrics.accuracy_score(y_test, y_pred6)
    print("Accuracy calculated for Class_label "+ class_label_iter + ": Model 6") 
    
    ####################################################
    accuracy_df.iloc[i,1] = accuracy_mod1
    accuracy_df.iloc[i,2] = accuracy_mod2
    accuracy_df.iloc[i,3] = accuracy_mod3
    accuracy_df.iloc[i,4] = accuracy_mod4
    accuracy_df.iloc[i,5] = accuracy_mod5
    accuracy_df.iloc[i,6] = accuracy_mod6
```

```python
accuracy_df.to_csv("./q3_output.csv")
```

### Conclusion:

### what’s going on as C increases for Linear models: From the accuracy_df values (q3_output.csv), we can say that accuracy remains almost same even with change in C values

### what’s going on as Kernel goes from Linear, Polynomial, and RBF for C = 5: From the accuracy_df values (q3_output.csv), we can say that accuracy increases when Kernel goes from linear (C=5), Polynomial (C=5, d=2) and RBF (C=5) whereas accuracy of Polynomial (C=5, D=3) is almost same as RBF
